Ejercicio 1. Seleccionar una API pública y extrae datos utilizando Python.

Utilizando la API de Pokemon para obtener los datos de cada pokemon usando el nombre para buscarlo.

```python
import requests

def get_pokemon_data(pokemon_name):
    # Fetch and print data for a given pokemon from PokeAPI, and download its sprite.
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        print_pokemon_summary(data)
        
    else:
        print(f"Failed to retrieve data for {pokemon_name}. Status code: {response.status_code}")

def print_pokemon_summary(data):
    #Print a summary of important pokemon data.
    name = data['name']
    poke_id = data['id']
    height = data['height']
    weight = data['weight']
    types = [t['type']['name'] for t in data['types']]

    print(f"Name: {name.capitalize()}")
    print(f"ID: {poke_id}")
    print(f"Height: {height}")
    print(f"Weight: {weight}")
    print("Types: " + ", ".join(types))

pokemon_name = "pikachu" 
get_pokemon_data(pokemon_name)
```

Ejercicio 2. Seleccionar un sitio web y extrae datos utilizando la técnica de Web Scraping en Python.

Usando el siguiente codigo para hacer Web Scraping en la pagina de Freshly Cosmetics (pagina española de productos de cosmetica natural), buscando obtener un listado de los links que tiene la pagina.  

```python
pip install beautifulsoup4
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.freshlycosmetics.com/es/"
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    
    links = []
    for a_tag in soup.find_all("a", href=True):  
        links.append(a_tag["href"])
    
    
    print("Extracted Links:")
    for link in links:
        print(link)
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
